```python
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# 1. Channel-wise NC Plot
# ==============================
attacks = ['No Attack', 'Gaussian Noise', 'Poisson Noise', 'Sharpening', 'Resizing']
r = [1.0, 0.9991, 0.9990, 1.0, 0.9882]
g = [1.0, 0.9989, 0.9988, 1.0, 0.9878]
b = [1.0, 0.9990, 0.9989, 1.0, 0.9876]

x = np.arange(len(attacks))
width = 0.25

plt.figure(figsize=(10,6))
plt.bar(x - width, r, width, label='R Channel')
plt.bar(x, g, width, label='G Channel')
plt.bar(x + width, b, width, label='B Channel')

plt.axhline(0.985, linestyle='--', color='black', label='Threshold')
plt.xticks(x, attacks)
plt.ylabel('NC per Channel')
plt.title('Channel-wise NC Under Various Attacks (R, G, B)')
plt.legend()
plt.show()


# ==============================
# 2. NC & PSNR vs Alpha
# ==============================
alpha = [5,10,15,20,25,30,35,40,50]
nc_no_attack = [0.92,0.962,0.982,0.995,0.999,1.0,1.0,1.0,1.0]
nc_gaussian = [0.89,0.93,0.96,0.982,0.997,0.999,0.999,1.0,1.0]
psnr = [43,38,34,31,29,27,25.5,24,22]

fig, ax1 = plt.subplots(figsize=(10,6))

ax1.plot(alpha, nc_no_attack, marker='o', label='NC (No Attack)')
ax1.plot(alpha, nc_gaussian, marker='s', linestyle='--', label='NC (Gaussian)')
ax1.set_xlabel('Embedding Strength (α)')
ax1.set_ylabel('NC')

ax2 = ax1.twinx()
ax2.plot(alpha, psnr, marker='^', linestyle=':', color='red', label='PSNR (dB)')
ax2.set_ylabel('PSNR (dB)')

ax1.legend(loc='lower right')
plt.title('NC and PSNR vs Embedding Strength (α)')
plt.show()


# ==============================
# 3. Radar Chart
# ==============================
labels = ['NC Robustness','PSNR Quality','Noise Resistance','Security Level','Blind Extraction','Comp. Efficiency']
values = [1.0,0.8,0.95,0.9,1.0,0.92]

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
values = np.concatenate((values,[values[0]]))
angles = np.concatenate((angles,[angles[0]]))

plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.3)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.title('Overall System Performance\n(Normalized Metrics)')
plt.show()


# ==============================
# 4. Filter Analysis
# ==============================
filters = ['No Filter','Gaussian σ=1','Gaussian σ=2','Sharpening (low)','Sharpening (high)','Median 3x3']
nc = [1.0,0.9995,0.998,1.0,0.9998,0.997]
psnr = [0,31,28,24,23,30]

x = np.arange(len(filters))
width = 0.35

fig, ax1 = plt.subplots(figsize=(10,6))
ax1.bar(x, nc, width, label='NC')
ax1.set_ylabel('NC')

ax2 = ax1.twinx()
ax2.bar(x, psnr, width, alpha=0.6, label='PSNR (dB)')
ax2.set_ylabel('PSNR (dB)')

plt.xticks(x, filters)
plt.title('Filter Analysis: NC and PSNR Under Filtering Attacks')
plt.show()


# ==============================
# 5. Noise Strength vs NC
# ==============================
sigma = [0,5,8,10,12,15,18,20,25,30]
gaussian_nc = [1.0,0.9999,0.9997,0.9995,0.9990,0.9982,0.997,0.996,0.9935,0.990]
poisson_nc = [1.0,0.9998,0.9996,0.9993,0.9989,0.9980,0.9965,0.995,0.992,0.9885]

plt.figure(figsize=(10,6))
plt.plot(sigma, gaussian_nc, marker='o', label='Gaussian Noise')
plt.plot(sigma, poisson_nc, marker='s', linestyle='--', label='Poisson Noise')

plt.axhline(0.985, linestyle=':', label='Robustness threshold')
plt.xlabel('Noise Strength (σ)')
plt.ylabel('Normalized Correlation (NC)')
plt.title('Noise Analysis: NC vs Noise Strength')
plt.legend()
plt.show()


# ==============================
# 6. PSNR Comparison
# ==============================
attacks = ['Gaussian Noise','Poisson Noise','Sharpening','Resizing']
psnr = [25.59,26.73,24.45,35.32]

plt.figure(figsize=(8,5))
bars = plt.bar(attacks, psnr)

plt.axhline(24, linestyle='--', label='Threshold (24 dB)')
plt.ylabel('PSNR (dB)')
plt.title('PSNR Comparison After Various Attacks')

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x()+0.1, y+0.3, f"{y:.2f} dB")

plt.legend()
plt.show()


# ==============================
# 7. NC Comparison
# ==============================
attacks = ['No Attack','Gaussian Noise','Poisson Noise','Sharpening','Resizing']
nc = [1.0,0.9990,0.9989,1.0,0.9878]

plt.figure(figsize=(8,5))
bars = plt.bar(attacks, nc)

plt.axhline(0.98, linestyle='--', color='red', label='Threshold (NC=0.98)')
plt.ylabel('Normalized Correlation (NC)')
plt.title('NC Comparison Across Attack Types')

for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x()+0.1, y+0.0005, f"{y:.4f}")

plt.legend()
plt.show()
```
